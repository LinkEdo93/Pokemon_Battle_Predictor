### Building a prediction model using Machine Learning.

```python
# First step, load battle data into the dataframe
df_battles = pd.read_csv(r'C:\Users\Edo\Documents\data_science-portfolio\Project1\data\final_combats.csv') # To load the data, change the path to correspond the file path in your device

# Display the first few rows of the data
df_battles.head()
```

```python
# Now, we have to merge the battle data with the pokemon data so we can get all the information related to the stats
# First, we merge PokÃ©mon data with battle data for the attacker PokÃ©mon
df_battles = df_battles.merge(df_pokemon, how='left', left_on='First_pokemon', right_on='#', suffixes=('_attacker', '_defender'))

# Second, we merge PokÃ©mon data with battle data for the defender PokÃ©mon
df_battles = df_battles.merge(df_pokemon, how='left', left_on='Second_pokemon', right_on='#', suffixes=('_attacker', '_defender'))

# Then we drop unnecessary columns such as original PokÃ©mon IDs
df_battles = df_battles.drop(['#_attacker', '#_defender'], axis=1)

# Display first few rows of the merged data
df_battles.head()
```

```python
# For the next step, we need to create features of all the stats for the machine learning model.
# To do so, we need to create stat difference features for all the stats between attacker and defender
df_battles['Attack_diff'] = df_battles['Attack_attacker'] - df_battles['Attack_defender']
df_battles['Defense_diff'] = df_battles['Defense_attacker'] - df_battles['Defense_defender']
df_battles['Speed_diff'] = df_battles['Speed_attacker'] - df_battles['Speed_defender']
df_battles['Sp_Atk_diff'] = df_battles['Sp. Atk_attacker'] - df_battles['Sp. Atk_defender']
df_battles['Sp_Def_diff'] = df_battles['Sp. Def_attacker'] - df_battles['Sp. Def_defender']
df_battles['HP_diff'] = df_battles['HP_attacker'] - df_battles['HP_defender']
```

```python
# We will also take into account both the 1st and 2nd type of the pokemons battling.
# To do so, fist we need to create a type match for both 1st and 2nd type
df_battles['Type1_matchup'] = (df_battles['Type 1_attacker'] == df_battles['Type 1_defender']).astype(int)
df_battles['Type2_matchup'] = (df_battles['Type 2_attacker'] == df_battles['Type 2_defender']).astype(int)
```

```python
# At this stage, we have included all the relevant data in order to build the model so we can proceed further
# First, let's define a function to calculate the type effectiveness for the battle
def calculate_type_effectiveness(attacker_type1, attacker_type2, defender_type1, defender_type2):
    # We calculate the effectiveness of the attacker's types against the defender's types
    effectiveness1 = type_effectiveness.get(attacker_type1, {}).get(defender_type1, 1) * type_effectiveness.get(attacker_type1, {}).get(defender_type2, 1)
    effectiveness2 = type_effectiveness.get(attacker_type2, {}).get(defender_type1, 1) * type_effectiveness.get(attacker_type2, {}).get(defender_type2, 1)
    
    # Finally, we combine the effectiveness scores
    combined_effectiveness = effectiveness1 * effectiveness2
    return combined_effectiveness
```

```python
# After defining our function, we will apply it to the battles dataset so we can calculate the type effectiveness for each battle
df_battles['Type_effectiveness'] = df_battles.apply(
    lambda row: calculate_type_effectiveness(
        row['Type 1_attacker'], row['Type 2_attacker'],
        row['Type 1_defender'], row['Type 2_defender']
    ), axis=1
)
```

```python
# Now we define the target variable as 1 if the attacker wins, or 0 if the defender wins
df_battles['Target'] = (df_battles['Winner'] == df_battles['First_pokemon']).astype(int)
```

```python
# the next step is to separate the features from the target and split the dataset into training and testing sets
from sklearn.model_selection import train_test_split

# Select features (X) and target (y)
X = df_battles[['Attack_diff', 'Defense_diff', 'Speed_diff', 'Sp_Atk_diff', 'Sp_Def_diff', 'HP_diff', 'Type1_matchup', 'Type2_matchup', 'Type_effectiveness']]  # Features
y = df_battles['Target']  # Target variable (attacker wins)

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=9)
```

```python
# Now, let's train a machine learning model using different models in order to find the most accurate
# Let's start with the Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

# First, we initialize the RandomForest model
model = RandomForestClassifier(n_estimators=100, random_state=9)

# Then, we train the model on the training data
model.fit(X_train, y_train)

# And finally, we evaluate the model on the test set
rf_accuracy = model.score(X_test, y_test)
print(f'Random Forest Classifier Accuracy: {rf_accuracy}')
```

```python
# Now, let's try Logistic Regression
from sklearn.linear_model import LogisticRegression
log_reg_model = LogisticRegression(random_state=9)
log_reg_model.fit(X_train, y_train)
y_pred_log_reg = log_reg_model.predict(X_test)
log_reg_accuracy = (y_pred_log_reg == y_test).mean()
print(f'Logistic Regression Accuracy: {log_reg_accuracy}')
```

```python
# Let's also try the Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(random_state=9)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)
dt_accuracy = (y_pred_dt == y_test).mean()
print(f'Decision Tree Accuracy: {dt_accuracy}')
```

```python
# Lastly, let's also try the Support Vector Classifier (SVC)
from sklearn.svm import SVC
svc_model = SVC(random_state=9)
svc_model.fit(X_train, y_train)
y_pred_svc = svc_model.predict(X_test)
svc_accuracy = (y_pred_svc == y_test).mean()
print(f'Support Vector Classifier Accuracy: {svc_accuracy}')
```

```python
# Finally, we'll compare them in order to find the model with the highest accuracy
models = {
    'Random Forest': rf_accuracy,
    'SVC': svc_accuracy,
    'Decision Tree': dt_accuracy,
    'Logistic Regression': log_reg_accuracy
}

# Print the model with the best accuracy
best_model_name = max(models, key=models.get)
print(f'Best Model: {best_model_name} with Accuracy: {models[best_model_name]}')
```

```python
import joblib

# Save the model to a file
joblib.dump(final_rf_model, 'rf_model.pkl')
```

## Finally, we have to deply the model using a Web API

```python
# First, let's import the required
from flask import Flask, request, jsonify
import threading

# Load the trained model
model = joblib.load('rf_model.pkl')

# Initialize Flask app
app = Flask(__name__)

# Define a route for prediction
@app.route('/predict', methods=['POST'])
def predict():
    # Get the data from the request
    data = request.get_json()
    features = np.array(data['features']).reshape(1, -1)  # Ensure the input is in the correct format

    # Make prediction
    prediction = model.predict(features)

    # Return the prediction as a JSON response
    return jsonify({'prediction': int(prediction[0])})  # Convert to int for JSON compatibility

# Run the app
def run_app():
    app.run(debug=True, use_reloader=False, port=5000)  # Port 5000 for local testing

# Run Flask app in a separate thread to avoid blocking Jupyter notebook
thread = threading.Thread(target=run_app)
thread.start()
```

```python
# Let's test it locally with some example features to see if the predictive model works as intended
import requests

# URL of the Flask app
url = 'http://127.0.0.1:5000/predict'

# Example data 
data = {
    'features': [100, 50, 25, 30, 40, 60, 1, 0, 0.75]
}

# Send the POST request
response = requests.post(url, json=data)

# Print the response from the Flask API
print(response.json())
```
